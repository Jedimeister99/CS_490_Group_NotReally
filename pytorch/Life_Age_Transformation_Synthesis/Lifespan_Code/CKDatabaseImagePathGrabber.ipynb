```python
from cv2 import cv2
import numpy as np
import os
```

```python
#Open the text file
textFileName = os.path.join("./males_image_list.txt")
File_object = open(textFileName,"w+")
```

```python
#Grab all the files from the last directory from every directory in CK+
CKfilePath = "/media/Data2/Cohn-Kanade/CK+/cohn-kanade-images"
CKimageName = ""
CKimageList = []
for subject in os.listdir(CKfilePath):
    if(os.path.isdir(os.path.join(CKfilePath, subject))):
        for seq in os.listdir(os.path.join(CKfilePath, subject)):
            if(os.path.isdir(os.path.join(CKfilePath, subject, seq))):
                CKtempList = []
                for image in os.listdir(os.path.join(CKfilePath, subject, seq)):
                    if(image.endswith(".png")):
                        CKtempList.append(image)
                CKtempList.sort()
                CKimageList.append(os.path.join(CKfilePath, subject,seq,CKtempList[-1]))
                File_object.write(os.path.join(CKfilePath, subject,seq,CKtempList[-1]))
                File_object.write("\n")
print(CKimageList)       
```

```python
File_object.close()
```
